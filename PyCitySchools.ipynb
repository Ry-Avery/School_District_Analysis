```python
# Importing dependencies
import pandas as pd
import os
```

```python
# Declaring variables to file path of data
school_data_to_load=os.path.join('Resources','schools_complete.csv')
student_data_to_load=os.path.join('Resources','students_complete.csv')
```

```python
# Creating DataFrame for school data
school_data_df=pd.read_csv(school_data_to_load)
school_data_df
```

```python
# Creating DataFrame for student data
student_data_df=pd.read_csv(student_data_to_load)
student_data_df.head(10)
```

```python
# Defining percentage() function
def percentage(Value,Total):
    return float(Value)/float(Total)*100
```

```python
# Looping through 'student_name' Series in student_data_df to replace all unwanted prefixes and suffixes
prefixes_suffixes=['Dr. ','Mr. ','Ms. ','Mrs. ','Miss ',' MD',' DDS',' DVM',' PhD']
for p_s in prefixes_suffixes:
    student_data_df['student_name']=student_data_df['student_name'].str.replace(p_s,'')
student_data_df.head(10)
```

```python
# Merging student_data_df and school_data_df into a new DataFrame 
school_data_complete_df=pd.merge(student_data_df,school_data_df,on='school_name')
school_data_complete_df.head()
```

```python
# Counting number of cells in the 'Student ID' column in school_data_complete_df
student_count=school_data_complete_df['Student ID'].count()
print(f'# of Students- {student_count:,}')
```

```python
# Getting number of unique cells in the 'school_name' column in school_data_complete_df
school_count=len(school_data_complete_df['school_name'].unique())
print(f'# of Schools- {school_count:,}')
```

```python
# Adding up the sum of cells in the 'budget' column in school_data_complete_df
total_budget=school_data_df['budget'].sum()
print(f'Total Budget- ${total_budget:,}')
```

```python
# Finding the average value of cells in the 'reading_score' column in school_data_complete_df
average_reading_score=school_data_complete_df['reading_score'].mean()
print(f'Average Reading Score- {average_reading_score:.1f}%')
```

```python
# Finding the average value of cells in the 'math_score' column in school_data_complete_df
average_math_score=school_data_complete_df['math_score'].mean()
print(f'Average Math Score- {average_math_score:.1f}%')
```

```python
# Finding number of cells in the 'reading_score' column in school_data_complete_df that are >=70
passing_reading_df=school_data_complete_df[school_data_complete_df['reading_score']>=70]
passing_reading_count=passing_reading_df['reading_score'].count()
print(f'# Passed Reading- {passing_reading_count:,}')
```

```python
# Finding number of cells in the 'math_score' column in school_data_complete_df that are >=70
passing_math_df=school_data_complete_df[school_data_complete_df['math_score']>=70]
passing_math_count=passing_math_df['math_score'].count()
print(f'# Passed Math- {passing_math_count:,}')
```

```python
# Getting the percentage of passing_reading_count divided by student_count with the percentage() function
passing_reading_percentage=percentage(passing_reading_count,student_count)
print(f'% Passed Reading- {passing_reading_percentage:.1f}%')
```

```python
# Getting the percentage of passing_reading_count divided by student_count with the percentage() function
passing_math_percentage=percentage(passing_math_count,student_count)
print(f'% Passed Math- {passing_math_percentage:.1f}%')
```

```python
# Finding number of rows that both 'reading_score' and 'math_score' columns in school_data_complete_df that are >=70
passing_both_df=school_data_complete_df[(school_data_complete_df['reading_score']>=70)&(school_data_complete_df['math_score']>=70)]
passing_both_count=passing_both_df['student_name'].count()
print(f'# Passed Both- {passing_both_count:,}')
```

```python
# Getting the percentage of passing_both_count divided by student_count with the percentage() function
passing_both_percentage=percentage(passing_both_count,student_count)
print(f'% Passed Both- {passing_both_percentage:.1f}%')
```

```python
# Creating DataFrame that contains school district summary
district_summary_df=pd.DataFrame(
    [{'Total Students':student_count,
    'Total Schools':school_count,
    'Total Budget':total_budget,
    'Average Reading Score':average_reading_score,
    'Average Math Score':average_math_score,
    '% Passing Reading':passing_reading_percentage,
    '% Passing Math':passing_math_percentage,
    '% Overall Passing':passing_both_percentage}])
district_summary_df
```

```python
# Formatting columns in district_summary_df
district_summary_df['Total Students']=district_summary_df['Total Students'].map('{:,}'.format)
district_summary_df['Total Budget']=district_summary_df['Total Budget'].map('${:,.2f}'.format)
district_summary_df['Average Reading Score']=district_summary_df['Average Reading Score'].map('{:.1f}'.format)
district_summary_df['Average Math Score']=district_summary_df['Average Math Score'].map('{:.1f}'.format)
district_summary_df['% Passing Reading']=district_summary_df['% Passing Reading'].map('{:.0f}%'.format)
district_summary_df['% Passing Math']=district_summary_df['% Passing Math'].map('{:.0f}%'.format)
district_summary_df['% Overall Passing']=district_summary_df['% Overall Passing'].map('{:.0f}%'.format)
district_summary_df
```

```python
# Renaming column names in district_summary_df
new_column_order=['Total Schools','Total Students','Total Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']
district_summary_df=district_summary_df[new_column_order]
district_summary_df
```

```python
# Creating a Series, setting its index to 'school_name' column, and adding 'type' column in school_data_df
school_types_s=school_data_df.set_index(['school_name'])['type']
school_types_s.head()
```

```python
# Creating a Series, setting its index to 'school_name' column, and adding 'size' column in school_data_df
school_counts_s=school_data_df.set_index(['school_name'])['size']
school_counts_s.head()
```

```python
# Creating a Series, setting its index to 'school_name' column, and adding 'budget' column in school_data_df
school_budget_s=school_data_df.set_index(['school_name'])['budget']
school_budget_s.head()
```

```python
# Creating a new Series by divding two existing Series
school_capita_s=school_budget_s/school_counts_s
school_capita_s.head()
```

```python
# Creating a Series, averaging 'math_score' column and grouping results by the 'school_name' column in school_data_df
school_math_s=school_data_complete_df.groupby(['school_name']).mean()['math_score']
school_math_s.head()
```

```python
# Creating a Series, averaging 'reading_score' column and grouping results by the 'school_name' column in school_data_df
school_reading_s=school_data_complete_df.groupby(['school_name']).mean()['reading_score']
school_reading_s.head()
```

```python
# Finding number of cells in the 'math_score' column and then 'reading_score' column in school_data_complete_df that are >=70
school_passing_math_df=school_data_complete_df[school_data_complete_df['math_score']>=70]
school_passing_reading_df=school_data_complete_df[school_data_complete_df['reading_score']>=70]
```

```python
# Creating a Series, counting 'student_name' column and grouping results by the 'school_name' column in school_data_df for both math and reading
school_passing_math_s=school_passing_math_df.groupby(['school_name']).count()['student_name']
school_passing_reading_s=school_passing_reading_df.groupby(['school_name']).count()['student_name']
```

```python
# Dividing Series by school_counts_s then multiplying by 100 to get a percentage
school_passing_math_s=school_passing_math_s/school_counts_s*100
school_passing_reading_s=school_passing_reading_s/school_counts_s*100
```

```python
# Finding number of rows that both 'reading_score' and 'math_score' columns in school_data_complete_df that are >=70
school_passing_both_df=school_data_complete_df[(school_data_complete_df['reading_score']>=70)&(school_data_complete_df['math_score']>=70)]
```

```python
# Creating a Series, counting 'student_name' column and grouping results by the 'school_name' column in school_data_df
school_passing_both_s=school_passing_both_df.groupby(['school_name']).count()['student_name']
```

```python
# Dividing Series by school_counts_s then multiplying by 100 to get a percentage
school_passing_both_s=school_passing_both_s/school_counts_s*100
school_passing_both_s
```

```python
# Creating DataFrame that contains school summary
school_summary_df=pd.DataFrame({
    'School Type':school_types_s,
    'Total Students':school_counts_s,
    'Total School Budget':school_budget_s,
    'Per Student Budget':school_capita_s,
    'Average Math Score':school_math_s,
    'Average Reading Score':school_reading_s,
    '% Passing Math':school_passing_math_s,
    '% Passing Reading':school_passing_reading_s,
    '% Overall Passing':school_passing_both_s})
school_summary_df.index.name='School Name'
school_summary_df.head()
```

```python
# Formatting columns in school_summary_df
school_summary_df['Total Students']=school_summary_df['Total Students'].map('{:,}'.format)
school_summary_df['Total School Budget']=school_summary_df['Total School Budget'].map('${:,.2f}'.format)
school_summary_df['Per Student Budget']=school_summary_df['Per Student Budget'].map('${:,.2f}'.format)
school_summary_df['Average Math Score']=school_summary_df['Average Math Score'].map('{:.1f}%'.format)
school_summary_df['Average Reading Score']=school_summary_df['Average Reading Score'].map('{:.1f}%'.format)
school_summary_df['% Passing Math']=school_summary_df['% Passing Math'].map('{:.1f}%'.format)
school_summary_df['% Passing Reading']=school_summary_df['% Passing Reading'].map('{:.1f}%'.format)
school_summary_df['% Overall Passing']=school_summary_df['% Overall Passing'].map('{:.1f}%'.format)
school_summary_df
```

```python
# Creating new DataFrame that sorts '% Overall Passing' column in school_summary_df from decending
top_schools_df=school_summary_df.sort_values(['% Overall Passing'],ascending=False)
top_schools_df.head()
```
